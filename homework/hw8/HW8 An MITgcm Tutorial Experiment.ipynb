### Part 4: Assess the Model Results
Upon successful completion of your model run, you should have the following files: `THETA.0000070445.data`, `SALT.0000070445.data`, `UVEL.0000070445.data`, and `VVEL.0000070445.data`. These files contain the surface fields for temperature, salinity, zonal velocity, and meridional velocity in the month of December. 

Read these files into four separate grids and make a 2x2 set of subplots showing the surface temperature, salinity, zonal velocity, and meridional velocity simulated by the model. 

````python
# read in the data from files into grids for theta, salt, uvel, and vvel

theta =  os.path.join(model_dir, 'run', 'THETA.0000070445.data')
theta_grid = np.fromfile(theta, '>f4').reshape((sNx, sNy))

salt =  os.path.join(model_dir, 'run', 'SALT.0000070445.data')
salt_grid = np.fromfile(salt, '>f4').reshape((sNx, sNy))

uvel =  os.path.join(model_dir, 'run', 'UVEL.0000070445.data')
uvel_grid = np.fromfile(uvel, '>f4').reshape((sNx, sNy))

vvel =  os.path.join(model_dir, 'run', 'VVEL.0000070445.data')
vvel_grid = np.fromfile(vvel, '>f4').reshape((sNx, sNy))
````

````python
# make the 2x2 plot here

fig = plt.figure(figsize = (17,10))
plt.subplot(2,2,1)
C = plt.pcolormesh(theta_grid, cmap='seismic')
plt.colorbar(C).set_label("tempurature")
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Surface temperature')
plt.subplot(2,2,2)
C = plt.pcolormesh(salt_grid, cmap='Blues')
plt.colorbar(C).set_label("salinity")
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Model Salinity')
plt.subplot(2,2,3)
C = plt.pcolormesh(uvel_grid, cmap='seismic')
plt.colorbar(C).set_label("zonal velocity")
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Zonal Velocity')
plt.subplot(2,2,4)
C = plt.pcolormesh(vvel_grid,cmap='seismic')
plt.colorbar(C).set_label("meridional velocity")
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Meridional Velocity')
plt.show()
````
